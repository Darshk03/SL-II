## 3.	Write a program to plot ReLu activation functions that are being used in neural networks.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def relu(x):
    return np.maximum(0,x)
```

```python
relu(3)
```

```python
xseriesrelu=np.linspace(-10,10,101)
xseriesrelu
```

```python
outputseriesrelu=relu(xseriesrelu)
outputseriesrelu
```

```python
plt.plot(xseriesrelu,outputseriesrelu,'go')
```

```python
plt.plot(xseriesrelu,outputseriesrelu,linewidth=5,color='y')
```
