```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def soft_max(x):
    return np.exp(x)/np.sum(np.exp(x))
```

```python
soft_max(5)
```

```python
xseriessoft=np.linspace(-10,10,101)
xseriessoft
```

```python
outputseriessoft=soft_max(xseriessoft)
outputseriessoft
```

```python
plt.plot(xseriessoft,outputseriessoft,'go')
```

```python
plt.plot(xseriessoft,outputseriessoft,linewidth=5,color='y')
```
