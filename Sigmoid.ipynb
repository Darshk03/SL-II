## 1. Write a program to plot Sigmoid activation functions that are being used in neural networks.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid_function(x):
    z=1/(1+np.exp(-x))
    return z
```

```python
sigmoid_function(5)
```

```python
xseries=np.linspace(-10,10,101)
xseries
```

```python
outputseries=sigmoid_function(xseries)
outputseries
```

```python
plt.plot(xseries,outputseries,'go')
```

```python
plt.plot(xseries,outputseries,linewidth=5,color='y')
```
