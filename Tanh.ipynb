## 2.	Write a program to plot Tanh activation functions that are being used in neural networks.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def tanh_function(x):
    return np.tanh(x)
```

```python
tanh_function(5)
```

```python
xseriestanh=np.linspace(-10,10,101)
xseriestanh
```

```python
outputseriestanh=tanh_function(xseriestanh)
outputseriestanh
```

```python
plt.plot(xseriestanh,outputseriestanh,'go')
```

```python
plt.plot(xseriestanh,outputseriestanh,linewidth=5,color='y')
```
